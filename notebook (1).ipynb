```python
# normalizing the rank of the customers
max_date = day_df['dteday'].max()

rfm_df = day_df.groupby('weekday').agg({
    'dteday': lambda x: (max_date - x.max()).days,
    'instant': 'count',
    'cnt': 'sum'
}).reset_index()

rfm_df.columns = ['customer_id', 'recency', 'frequency', 'monetary']
```

```python
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['frequency'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_rank_norm'] = (rfm_df['f_rank']/rfm_df['f_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100

rfm_df.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)

rfm_df['RFM_score'] = 0.15*rfm_df['r_rank_norm'] + 0.28*rfm_df['f_rank_norm'] + 0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)

print("\nRFM DataFrame (First 7):")
print(rfm_df[['customer_id', 'RFM_score']].head(7))
print("\nFull RFM Table:")
print(rfm_df)
```

#### hasil segmentasi penggunaan sepeda berdasarkan weekday (hari dalam seminggu) yang diurutkan dari skor tertinggi

```python
max_date = day_df['dteday'].max()

rfm_df = day_df.groupby('weekday').agg({
    'dteday': lambda x: (max_date - x.max()).days, # Recency
    'instant': 'count',                             # Frequency
    'cnt': 'sum'                                   # Monetary
}).reset_index()

rfm_df.columns = ['weekday', 'recency', 'frequency', 'monetary']

# Assign scores (1-5)
# Since we only have 7 records (0-6), quantiles might be tricky with q=5.
# We'll use rank or simple linear scaling/binning if possible, but let's try rank-based quantiles first.
# If values are too similar, we'll handle that.

def get_score(series, ascending=True):
    # Using rank to avoid errors with identical values in quantiles
    ranks = series.rank(method='first', ascending=ascending)
    return pd.cut(ranks, bins=5, labels=[1, 2, 3, 4, 5]).astype(int)

# Recency: Lower is better (score 5)
rfm_df['r_score'] = get_score(rfm_df['recency'], ascending=False)
# Frequency: Higher is better (score 5)
rfm_df['f_score'] = get_score(rfm_df['frequency'], ascending=True)
# Monetary: Higher is better (score 5)
rfm_df['m_score'] = get_score(rfm_df['monetary'], ascending=True)

# Calculate average RFM Score
rfm_df['RFM_score'] = rfm_df[['r_score', 'f_score', 'm_score']].mean(axis=1)
```

```python
rfm_df["customer_segment"] = np.where(
    rfm_df['RFM_score'] > 4.5, "Top customers", (np.where(
        rfm_df['RFM_score'] > 4, "High value customer",(np.where(
            rfm_df['RFM_score'] > 3, "Medium value customer", np.where(
                rfm_df['RFM_score'] > 1.6, 'Low value customers', 'lost customers'))))))

# Preparing for visualization
# 'customer_id' in user code is 'weekday' in our context
customer_segment_df = rfm_df.groupby(by="customer_segment", as_index=False).weekday.nunique()
customer_segment_df.rename(columns={'weekday': 'weekday_count'}, inplace=True)
```

```python
customer_segment_df['customer_segment'] = pd.Categorical(customer_segment_df['customer_segment'], [
    "lost customers", "Low value customers", "Medium value customer",
    "High value customer", "Top customers"
])
```

```python
plt.figure(figsize=(10, 5))
colors_ = ["#D3D3D3", "#D3D3D3", "#D3D3D3", "#72BCD4", "#72BCD4"] # Adjusted colors based on order

sns.barplot(
    x="weekday_count",
    y="customer_segment",
    data=customer_segment_df.sort_values(by="customer_segment", ascending=False),
    palette=colors_
)
plt.title("Number of Weekdays for Each Segment", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel("Count of Weekdays")
plt.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.savefig('weekday_segmentation.png')

print(rfm_df[['weekday', 'RFM_score', 'customer_segment']].sort_values(by='RFM_score', ascending=False))
```
